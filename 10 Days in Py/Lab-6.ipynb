**map function**

```python
lis1 = [8, 4, 6, 10, 9]
lis2 = [2, 3, 1, 2, 5]

def addition(x, y):
  return x+y

def subtraction(x, y):
  return x-y

list3 = list(map(addition, lis1, lis2))
list4 = list(map(subtraction, lis1, lis2))
print(list3)
print(list4)
```

```python
lis1 = [8, 4, 6, 10, 9]
lis2 = [2, 3, 1, 2, 5]

def calculation(x, y):
  return x+y, x-y

list(map(calculation, lis1, lis2))

```

**Finding any index and its value from a 2D List**

```python
lis = [[None, 'Research', 'Study', 'Game'], ['Naim', 2, 3, 4], ['Shuvo', 1, 1, 2]]
lis
```

```python
for i, row in enumerate(lis):
  for j, item in enumerate(row):
    print(f'a[{i}][{j}] = {item}', end=' ')
  print()
```
