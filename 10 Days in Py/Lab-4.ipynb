```python
lis = [3, 7, 9, 1, 5, 10]

lis2 = [item for item in reversed(lis)]
lis2
```

```python
numbers = [2, 3, 5, 6, 8, 9, 4]
for value in (x**2 for x in numbers if x%2 == 0):
  print(value, end=' ')
```

```python
lis = [[4, 5, 6], [1, 2, 3]]
lis
```

```python
for row in lis:
  for item in row:
    print(item, end=' ')
  print()
```

```python
for i, row in enumerate(lis):
  for j, item in enumerate(row):
    print(f'lis[{i}][{j}] = {item}' , end=' ')

  print()
```
