**Suppose you are going to build a system for a Kacchi-Bhai. Where customer can see the price and order two types of biriyani. One is Chicken Biriyani for 350 Taka and another one is Beef Biriyani for 450 Taka.**

*   Take input from user how many biriyani they want to order on each item.
*   Customer can order extra plate of rice for 100 Taka.
*   If they order more than two biriyanis, they will get 3% discount on total price.
*   Calculate the total price and show it to the customer.






```python
print("You can order Chicken Biriyani for 350 Taka and Beef Biriyani for 450 Taka")
a = int(input("How many Chicken Biriyani do you want?"))
b = int(input("How many Beef Biriyani do you want?"))
print("You can order extra plate of Rice for 100 Taka")
c = int(input("How many extra plate of Rice do you want?"))

total= a*350+b*450+c*100

if a+b>2:
  total = total - total*0.03
  print(f'Your total Bill with discount is {total} Taka only!')
else:
  print(f'Your total Bill is {total} Taka only!')
```
