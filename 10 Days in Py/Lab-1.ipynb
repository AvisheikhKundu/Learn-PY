**if-elif-else Statement**

```python
grade = int(input())
if grade >= 60:print("Passed")
else:print("Failed")
```

```python
grade = int(input("Enter your mark: "))
if grade>=90:
  print('"A"')
elif grade>=80:
  print('"B"')
elif grade>=70:
  print('"C"')
elif grade>=60:
  print('"D"')
else:
  print('"F"')
```
