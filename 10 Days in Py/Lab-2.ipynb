**Control Statement**

 A class of 10 students took a quiz. Their grades (integers in the range 0 â€“ 100) are 98, 76, 71, 87, 83, 90, 57, 79, 82, 94. Determine the class average on the quiz.

```python
total = 0
grades = [98, 76, 71, 87, 83, 90, 57, 79, 82, 94]

for grade in grades:
  total = total+grade

average = total/len(grades)
print(average)
print(f'Average is {average}')
```

**Sentinel-Controlled Solving**

```python
total = 0
counter = 0

grade = int(input("Enter grade, -1 to end: "))
while grade != -1:
  total = total+grade
  counter = counter+1
  grade = int(input("Enter grade, -1 to end: "))

average = total/counter
print(average)
print(f'Average is {average}')
```

```python
total = 0
counter = 0

mark = int(input("Enter mark, -1 to end: "))
while mark != -1:
  total = total+mark
  counter = counter+1
  mark = int(input("Enter mark, -1 to end: "))

average = total/counter
print(f'Average mark is {average}')
```

**Nested Control Statement**

*  A college offers a course that prepares students for the IELTS exam.
*  Last year, several of the students who completed this course took the IELTS
exam.
*  The college wants to know how well their students did on the exam.
*  You have been asked to write a program to summarize the results.
*  You have been given a list of 10 students.
*  Next to each name is written a 1 if the student passed the exam and a 2 if
the student failed.
*  If more than six students passed the exam, the instructor can get bonus.



```python
passes = 0
fails = 0

for number in range(10):
  result = int(input("Enter result: 1 for Pass and 2 for Fail: "))

  if result==1:
    passes = passes+1
  else:
    fails = fails+1

print(f'Total {passes} students passed and {fails} failed!')

if passes > 6:
  print("Teacher will get bonus!")
else:
  print("Teacher will not get bonus!")
```
